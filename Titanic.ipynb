```python
# Import the necessary modules
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# Load your dataset (CSV file)
x=data_numeric.drop(['Survived'], axis=1).values #data without surviived data
y=data_numeric['Survived'].values                #data of survived only.
```

```python
# Split the dataset into training and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python

from sklearn.metrics import accuracy_score
```

```python
# Create a logistic regression model
xy = LogisticRegression()

# Train the model on the training data
xy.fit(x_train,y_train)


# Evaluate the model on the testing data
xypred = xy.predict(x_test)
```

```python
accuracy_score(y_test,xypred)
```

```python
# Import the necessary modules
from sklearn.model_selection import GridSearchCV

c_space = np.logspace(-10, 10, 20)
param_grid = {'C':c_space}

# Create a GridSearchCV object with 5-fold cross-validation
logreg_cv = GridSearchCV(xy, param_grid, cv=5)

# Fit the GridSearchCV object to the training data
logreg_cv.fit(x_train,y_train)

# Print the tuned logistic regression parameters and best score
print("Turend Logistic Regression Parameters: {}",format(logreg_cv.best_params_))
print("Best Score is {}",format(logreg_cv.best_score_))
```
